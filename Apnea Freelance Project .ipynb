```python
mat3 = sio.loadmat('03_wsignevents_aligned.mat')
```

```python
#Call the Variables
Abdomen3 = mat3.get("Abdomen")
Abdomen3 = np.transpose(Abdomen3)
DruckFlow3 = mat3.get("DruckFlow")
H3 = mat3.get("ECG")
FlowTh3 = mat3.get("FlowTh")
Oesoph3 = mat3.get("Oesoph")
SPO23 = mat3.get("SPO2")
c_apnea3 = mat3.get("c_apnea")
hypopnea3 = mat3.get("hypopnea")
m_apnea3 = mat3.get("m_apnea")
o_apnea3 = mat3.get("o_apnea")
```

```python
#Convert them to lists 
One3 = list(DruckFlow3.flatten())
Two3 = list(H3.flatten())
Three3 = list(FlowTh3.flatten())
Four3 = list(Oesoph3.flatten())
Five3 = list(Abdomen3.flatten())
Six3 = list(SPO23.flatten())
Seven3 = list(c_apnea3.flatten())
Eight3 = list(hypopnea3.flatten())
Nine3 = list(m_apnea3.flatten())
Ten3 = list(o_apnea3.flatten())
```

```python
df3 = pd.DataFrame(
    {'DruckFlow': One3,
     'H': Two3,
     'FlowTh': Three3,
     'Oesoph': Four3,
     'Abdomen': Five3,
     'SPO2': Six3,
     'c_apnea': Seven3,
     'hypopnea': Eight3,
     'm_apnea': Nine3,
     'o_apnea': Ten3
    })
```

```python
df3
```

```python
#Function that creates True if ANY of the apnea columns is 0

df3['0Fault'] = df3['c_apnea'].apply(lambda x: True if (x == 0)  else False)
df3['1Fault'] = df3['hypopnea'].apply(lambda x: True if (x == 0)  else False)
df3['2Fault'] = df3['m_apnea'].apply(lambda x: True if (x == 0)  else False)
df3['3Fault'] = df3['o_apnea'].apply(lambda x: True if (x == 0)  else False)
```

```python
#Function that creates 0 or 1 depending on the output of the function above 

def f(row):
    if row['0Fault'] == True or row['1Fault'] == True or row['2Fault'] == True or row['3Fault'] == True :
        val = 0
    else:
        val = 1
    return val
```

```python
#Add a new column of the COMBINED result 0 or 1
df3['Label'] = df3.apply(f, axis=1)
```

```python
#Call the new DataFrame
df3
```

```python
#Drop Unwanted Columns 
df3 = df3.drop(['c_apnea', 'hypopnea','m_apnea','o_apnea','0Fault','1Fault','2Fault','3Fault'], axis=1)
```
